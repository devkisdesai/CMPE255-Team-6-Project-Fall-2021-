```python
# Import pacakages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pycountry
import re
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import cufflinks as cf
import geopandas as gpd
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# load data
data_1960_2050 = pd.read_csv('./Datasets/Minimized Data 1960-2050.csv', low_memory=False)
```

```python
# Removing Duplicated Fields
pop_data = data_1960_2050

i=1960
string =["Country_Name", "Country_Code", "Series_Name", "Series_Code"]
while i != 2051:
    temp = "[" + str(i) + "]"
    string.append(temp)
    i = i + 1
    
pop_data.columns = string

# replace .. with na
pop_data = pop_data.replace('..', np.nan)
pop_data = pop_data.set_index('Country_Name')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(pop_data.isnull().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.show()
```

```python
pop_data.head()
```

```python
#list of country codes
country_codes = []
country_names = []
for country in pycountry.countries:
    country_codes.append(country.alpha_3)
    country_names.append(country.name)
country_codes.sort(key=str.lower)
country_names.sort(key=str.lower)
```

```python
#countries data
countries_data = pop_data[pop_data.Country_Code.isin(country_codes)]
non_country_data = pop_data[~pop_data.Country_Code.isin(country_codes)]
# print(set(countries_data.Country_Name))
# print(set(non_country_data.Country_Name))
# no_data = (set(country_codes)) - set(countries_data.Country_Code)
```

```python
#segment data

#totals
pop_migration = countries_data[countries_data.Series_Code == 'SM.POP.NETM'] #net migration
pop_total = countries_data[countries_data.Series_Code == 'SP.POP.TOTL'] #total population
pop_rural = countries_data[countries_data.Series_Code == 'SP.RUR.TOTL'] #rural population
pop_urban = countries_data[countries_data.Series_Code == 'SP.URB.TOTL'] #urban population

#world data
world_total = non_country_data[(non_country_data.Series_Code == "SP.POP.TOTL") & (non_country_data.Country_Code == "WLD")]
world_growth = non_country_data[(non_country_data.Series_Code == "SP.POP.GROW") & (non_country_data.Country_Code == "WLD")]
#ratios
pop_birth_rate = countries_data[countries_data.Series_Code == 'SP.DYN.CBRT.IN'] #Birth rate, crude (per 1,000 people)
pop_death_rate = countries_data[countries_data.Series_Code == 'SP.DYN.CDRT.IN'] #Death rate, crude (per 1,000 people)
pop_male = countries_data[countries_data.Series_Code == 'SP.POP.TOTL.MA.ZS'] #Population, male (% of total population)
pop_female = countries_data[countries_data.Series_Code == 'SP.POP.TOTL.FE.ZS'] #Population, female (% of total population)
pop_age_dependency_w = countries_data[countries_data.Series_Code == 'SP.POP.DPND'] #Age dependency ratio (% of working-age population)
pop_age_dependency_o = countries_data[countries_data.Series_Code == 'SP.POP.DPND.OL'] #Age dependency ratio, old
pop_age_dependency_y = countries_data[countries_data.Series_Code == 'SP.POP.DPND.YG'] #Age dependency ratio, young
```

```python
#format data for visualizations
def format_data(df, names):
    columns = df.columns[4:]
    df = df.melt(value_vars=columns)
    df = df.rename(columns={"variable": names[0], "value": names[1]})
    df[names[0]] = df[names[0]].str.replace("[","")
    df[names[0]] = df[names[0]].str.replace("]","")
    df[names[1]]  = pd.to_numeric(df[names[1]], errors='coerce').fillna(0).astype(np.float64)
    return df

def preparePlot(dataframe): #We can use this method to plot individual plots (Can be expanded to be more automated)
    string = dataframe.index.tolist()
    dataframe = dataframe.T
    dataframe = dataframe.iloc[4:]
    dataframe = dataframe.astype('float64')
    dataframe.columns = string
    return dataframe


def pop_over_time_plot(country):
    pop_total_plot.loc[country].iplot(
                    xTitle='Year', 
                    yTitle='Population of {}'.format(country),
                    title='Population Over Time'
                     )
    
def pop_rural_over_time_plot(country):
    pop_rural_plot.loc[country].iplot(
                    xTitle='Year', 
                    yTitle='Population of {}'.format(country),
                    title='Rural Population Over Time'
                     )

def pop_urban_over_time_plot(country):
    pop_urban_plot.loc[country].iplot(
                    xTitle='Year', 
                    yTitle='Population of {}'.format(country),
                    title='Urban Population Over Time'
                     )
  
```

```python
#world pop est
world_gpd = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world_gpd = world_gpd.merge(pop_total, left_on="iso_a3", right_on="Country_Code")

world_gpd.plot(column='pop_est', cmap='OrRd', legend=True);
```

```python
# General visualizations - world population vs growth rate
world_total = format_data(world_total, ['year','population'])
world_growth = format_data(world_growth, ['year','growth'])
world_data = pd.merge(world_total, world_growth, on="year")

fig, ax = plt.subplots()

ax.plot(world_data['year'], world_data['population'], color='red')
ax.tick_params(axis='y', labelcolor='red')
ax2 = ax.twinx()

ax2.plot(world_data['year'], world_data['growth'], color='green')
ax2.tick_params(axis='y', labelcolor='green')
fig.autofmt_xdate(rotation=45)
ax.set_xticks(ax.get_xticks()[::5])
plt.show()
```

```python
# General visualizations - Migration Trends
migration_plot = preparePlot(pop_migration)
fig, axs = plt.subplots(figsize=(12, 4))  
migration_plot.plot.area(ax = axs, stacked = False)
axs.get_legend().remove()
axs.set_xlabel("Census Years")
axs.set_ylabel("Migration Trends")
axs.set_title("Net Migration by Country")
fig.savefig("Images/Migration Trends")
```
